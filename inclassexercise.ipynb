```python
import pandas as pd
```

```python
apr_2020p = pd.read_csv("time_use_2020-04.csv")
apr_2020p
```

```python
apr_2020p.dtypes
```

```python
apr_2020p["geslacht"]= apr_2020p["geslacht"].astype('category')
apr_2020p.dtypes
```

```python
apr_2020p = apr_2020p.rename(columns = {"geslacht": "gender",
        "v1q1_v1col1": "wh",
        "v1q1a_v1col1": "wh_hnk",
        "v1q1b_v1col1": "wh_hrk",
        "v1q1c_v1col1": "wh_hnrk",
        "v1q5_v1col1": "childcare_hours",
        "v1q5a_v1col1": "homeschooling_hours",
        "nohouse_encr": "hh_id",
        "nomem_encr": "ind_id"}) 
apr_2020p
```

```python
apr_2020p["month"]= "202004"
```

```python
nov_2019 = pd.read_stata("time_use_2019-11.dta", convert_categoricals=True)

nov_2019 = nov_2019.rename(
    columns={
        "geslacht": "gender",
        "v1q1_v1col1": "wh",
        "v1q5_v1col1": "childcare_hours",
        "nohouse_encr": "hh_id",
        "nomem_encr": "ind_id"
    }
)
nov_2019["ind_id"] = nov_2019["ind_id"].astype(int)
nov_2019["hh_id"] = nov_2019["hh_id"].astype(int)

nov_2019["month"]="201911"
nov_2019.dtypes
```

```python
df = apr_2020p.append(nov_2019)
```

```python
df_gender = df[["ind_id","gender"]]
df_gender
```

```python
df_gender = df_gender.drop_duplicates(keep="first")
df_gender
```

```python
df["new_id"] = df["ind_id"].astype(str) + df["month"].astype(str)
df.set_index("new_id",inplace=True)
df
```
